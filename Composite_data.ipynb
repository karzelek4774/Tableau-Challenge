```python
import pandas as pd
import glob
path = r"C:\Users\User\Documents\Bootcamp\Assignments\Homework\Tableau\data"
all_files = glob.glob(path + "\*.csv")
composite = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    composite.append(df)
composite_data = pd.concat(composite, axis=0, ignore_index=True)

composite_data.head()
```

```python
composite_data.to_csv("data/composite_data.csv", index=False, header=True)
```
